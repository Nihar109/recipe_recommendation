```python
import bs4
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
import pymongo
from pymongo import MongoClient
```

```python
url = 'https://www.bawarchi.com/recipe/11'
page = requests.get(url)
soup = bs(page.content,'html.parser')

count = 0
recipe_name = []
cusine = []
diet = []

for i in soup.findAll('a',class_='gallery'):
    #print(i['href'])
    page = bs(requests.get(i['href']).content)
    for title in page.find_all('title'):
        text1 = title.text
        var = text1.split(" | ")
        recipe_name.append(var[0][2:])
        cusine.append(var[1])
        diet.append(var[2])
        
data = {'recipe_name': recipe_name,'cusine':cusine,'diet':diet}
df = pd.DataFrame(data)            
```

```python
df = pd.DataFrame(data)            
```

```python
connect = MongoClient(host="localhost", port=27017)
database = connect["foodDB"]
collection = database["Recipe"]
rec_id = collection.insert_one(data)


# Printing the data inserted
cursor = collection.find()
for record in cursor:
    print(record)
```

```python
df.drop('_id', axis=1, inplace=True)
```

```python
df['cusine'] = df['cusine'].apply(lambda x: x.lower())
df['diet'] = df['diet'].apply(lambda x: x.lower())
```

```python
df['index'] = df.index
df['cusine_diet'] = df['cusine'] + df['diet']
df
```

```python
new_df = df[['index','recipe_name','cusine_diet']]
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words="english")
```

```python
vectors = cv.fit_transform(new_df['cusine_diet']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
def recommend(recipe):
    """Creating recommendation function to 
       recommend most five recipe_name from 
       user input
    """
    recipe_index = new_df[new_df['recipe_name'] == recipe].index[0]
    distances = similarity[recipe_index]
    recipe_list = sorted(enumerate(distances),reverse=True, key = lambda x:x[1])[1:6]
    
    for i in recipe_list:
        print(new_df.iloc[i[0]].recipe_name)
```

```python
recommend('Ras Malai with Eggs')
```
